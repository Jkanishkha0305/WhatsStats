```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/Users/kanishkhajaisankar/Downloads/WhatsStats/_chat.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    chat_data = file.read()

# Define a regex pattern to extract names
name_pattern = re.compile(r'\[(\d{2}\/\d{2}\/\d{4}, \d{2}:\d{2}:\d{2})\] (\w+):')

# Use the regex pattern to find all name occurrences in the chat
names = re.findall(name_pattern, chat_data)
```

```python
df = pd.DataFrame(names, columns=['Timestamp', 'Name'])
name_counts1 = df['Name'].value_counts()
print("Message count")
print(name_counts1)
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
name_counts1.plot(kind='bar')
plt.xlabel('Names')
plt.ylabel('Message Count')
plt.title('Ashooo grp Message count vs Names')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.tight_layout()

# Save the bar graph as an image (optional)
plt.savefig('message_count_bar_graph.png')

# Show the bar graph
plt.show()
```

```python
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
name_counts1.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Ashoo grp Message count Pie chart ')
plt.ylabel('')  # Remove the default ylabel
plt.tight_layout()

# Save the pie chart as an image (optional)
plt.savefig('message_distribution_pie_chart.png')

# Show the pie chart
plt.show()
```
